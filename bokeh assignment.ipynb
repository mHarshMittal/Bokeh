Q1. How can you create a Bokeh plot using Python code?

```python
# import modules
from bokeh.plotting import figure, output_notebook, show

# output to notebook
output_notebook()

# create figure
p = figure(outer_width = 400, outer_height = 400)

# add a circle renderer with
# size, color and alpha
p.circle([1, 2, 3, 4, 5], [4, 7, 1, 6, 3],
		size = 10, color = "navy", alpha = 0.5)

# show the results
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Glyphs in Bokeh are the visual shapes or markers that represent the data in a plot. Bokeh provides various types of glyphs, such as circles, squares, lines, and more. You can add glyphs to a Bokeh plot by calling the appropriate glyph methods on a figure object.

```python
from bokeh.plotting import figure, show

p = figure()

# Example data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Add a line glyph
p.line(x, y)

show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

To customize the appearance of a Bokeh plot, including the axes, title, and legend, you can use various attributes and methods provided by Bokeh.

```python
from bokeh.plotting import figure, show

p = figure(title="Customized Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Example data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Add a circle glyph
p.circle(x, y, size=10, color="red", legend_label="Data")

# Customize axes
p.xaxis.axis_label_text_font_style = "italic"
p.yaxis.axis_label_text_color = "blue"

# Customize title
p.title.text_color = "green"
p.title.text_font_size = "20pt"

show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A Bokeh server is a Python process that allows you to create and serve interactive Bokeh plots that can be updated in real time. The Bokeh server provides a way to execute Python callbacks in response to UI and tool events, enabling dynamic and interactive data visualization.

```python

from bokeh.plotting import figure, output_file, show

output_file("abc.html")


graph = figure(title = "Bokeh Line Graph")

# the points to be plotted
x = [1, 2, 3, 4, 5]
y = [1, 6, 8, 2, 3]

# plotting the line graph
graph.line(x, y)

# displaying the model
show(graph)
```
